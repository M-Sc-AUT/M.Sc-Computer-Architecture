```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 10, 0.0001)
A = 2
landa = 0.5
```

```python
y1 = A * np.exp(-landa * t)
y2 = 1 - y1
```

```python
plt.plot(t, y1, label="y1")
plt.plot(t, y2, label="y2")
plt.axhline(y = 0, color = 'red', linestyle = '--', label="y = 0") 
plt.axhline(y = 1, color = 'black', linestyle = '--', label="y = 1") 
plt.legend()

plt.show()
```

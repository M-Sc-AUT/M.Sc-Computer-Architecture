```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# img_array = np.loadtxt("MNIST/src/Data/output/MNIST_Out_HLS_1.txt", dtype=np.uint8)
# img = Image.fromarray(img_array)
# print("image shape: {}" .format(img_array.shape))

filenames = [
    "MNIST/src/Data/output/MNIST_Out_HLS_0.txt",
    "MNIST/src/Data/output/MNIST_Out_HLS_1.txt",
    "MNIST/src/Data/output/MNIST_Out_HLS_2.txt",
    "MNIST/src/Data/output/MNIST_Out_HLS_3.txt",
    "MNIST/src/Data/output/MNIST_Out_HLS_4.txt",
    "MNIST/src/Data/output/MNIST_Out_HLS_5.txt",
    "MNIST/src/Data/output/MNIST_Out_HLS_6.txt",
    "MNIST/src/Data/output/MNIST_Out_HLS_7.txt",
    "MNIST/src/Data/output/MNIST_Out_HLS_8.txt",
    "MNIST/src/Data/output/MNIST_Out_HLS_9.txt"
]
```

```python
fig, axes = plt.subplots(2, 5, figsize=(15, 5))

for i, filename in enumerate(filenames):
    img_array = np.loadtxt(filename, dtype=np.uint8)
    img = Image.fromarray(img_array)

    row = i // 5
    col = i % 5

    axes[row, col].imshow(img, cmap="gray")
    axes[row, col].axis("off")
    axes[row, col].set_title(f"Image {i + 1}")

plt.tight_layout()
plt.show()
```
